 #### **Importing and Viewing The Dataset**

 #### **Load The Dataset** 

 

```python
#Import the required libraries
import pandas as pd
import numpy as np

# Display dataset for viewing
data = pd.read_csv('../data/raw_data/SleepLifestyle.csv')
data
```

#### **Check Data Types**

```python
# Check the column names and data types
data.info()
```

All the dataset collumns are fully populated except "Sleep Disorder" so I will replace the missing values with "None".

```python
data.columns = data.columns.str.replace(" ", "_")
data["Sleep_Disorder"] = data["Sleep_Disorder"].fillna("None")
```

#### **Locate Missing Values**

```python
data.isnull().sum()
```

An alternative way of locating missing values can done by using a heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(data.isnull(), cbar=False)
plt.show
```

If the heatmap shows colours then it means that there are missing values and if it shows darkness, like this heatmap does, it means there are no missing values. 

#### **Check For Duplicates**

```python
data.duplicated().sum()
```

There are no duplicates in the dataset.

#### **Statistical Summary**

```python
data.describe()
```

The statistics of the dataset are up to standards as the Standard deviation (std) of each value is low which means that the data points are close to the averages; values are consistent.
